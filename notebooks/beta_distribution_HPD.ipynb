```julia
include("../src/Distribution/beta.jl")
using Plots
```

```julia
# 最高事後密度区間(HPD)の産出と描画
let
    a = 4
    b = 9
    αs = [0.01, 0.05, 0.1]
    o_plot = plot(0:0.01:1.0, x->beta_pdf(x,a,b), label="Beta(a=$a,b=$b)", linecolor = :darkgray)
    for α in αs
        hpd = beta_hpd(α, a, b)
        println(hpd)
        plot!([hpd[1], hpd[1]              , hpd[2]              , hpd[2]]
            , [0     , beta_pdf(hpd[1],a,b), beta_pdf(hpd[2],a,b), 0]
            , label="HPD α=$α")
    end
    plot(o_plot, fmt=:png)
end
```
