# 2次元確率ベクトルの正規分布
平均ベクトル $ \textbf{a} \in \mathbb{R}^2 $, 分散共分散行列 $ \Sigma $ の正規分布
$$
p(x) = \frac{1}{(2\pi)^{N/2}det(\Sigma)^{1/2}} \exp{\left(-\frac{(x-a)^{T} \Sigma^{-1} (x-a)}{2} \right)}
$$

下記は $ N = 2 $ の例である。

共分散行列は
$$
\Sigma =
\begin{pmatrix}
\sigma_{x}^2 & cov_{x,y} \\
cov_{x,y} & \sigma_{y}^2 \\
\end{pmatrix}
$$
で、共分散 $ cov_{x,y} $ が大きいほど、斜めになる。

```julia
# 平均ベクトル
mean_x = 1
mean_y = -0.5
mean_vector = [mean_x, mean_y]
# 分散共分散行列
Σ = [5.0 2.0
     ;2.0 3.0]
# 分散共分散行列の行列式および逆行列
detΣ = det(Σ)
invΣ = inv(Σ)

function norm_dist(x)
    a = (2π) * sqrt(detΣ)
    b = -(transpose(x - mean_vector)*invΣ*(x - mean_vector))/2
    return 1/a*exp(b)
end

contour(-5:0.05:5, -5:0.05:5, (x,y)->norm_dist([x,y]), fill = true, fmt=:png)
plot!([mean_x, mean_x], [-5, 5], label="maen(x)", linecolor=:lime)
plot!([-5, 5], [mean_y, mean_y], label="mean(y)", linecolor=:magenta)
```
