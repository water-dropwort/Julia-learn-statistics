```julia
using Plots
include("../src/Distribution/binomial.jl")
include("../src/Distribution/poisson.jl")
include("../src/Distribution/pascal.jl")
```

```julia
# 2項分布
let
    p = plot([],[],label="")
    for p in [0.2,0.8], N in [5,10]
        plot!(0:N, map(x->binom_dist(N,x,p), 0:N), label="p=$p N=$N")    
    end
    plot(p)
end
```

```julia
# ポアソン分布
let
    x_lim=15
    p = plot([],[],label="")
    for λ in [0.3, 2.0, 5.0]
        plot!(0:x_lim, map(x->poisson_dist(λ,x),0:x_lim), label="λ=$λ")
    end
    plot(p)
end
```

```julia
# パスカル分布
let
    x_lim = 25
    p = plot([],[],label="")
    for r in [1,3,5,7]
        # r=1は幾何分布
        plot!(r:x_lim, map(x->pascal_dist(x,r,0.5), r:x_lim), label="r=$r")
    end
    plot(p)
end
```
